```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
filename = 'data/table_i702t60.csv'
df = pd.read_csv(filename)

if not os.path.exists('images'):
    os.makedirs('images')
```

```python
df.head()
```

```python
month_number = df.loc[:,'month'].values
interest_paid = df.loc[:,'interest_paid'].values
principal_paid = df.loc[:,'principal_paid'].values
```

```python
plt.style.use('seaborn-v0_8')
```

```python
plt.figure(figsize=(10,5))
plt.plot(month_number,interest_paid,c='k',label="Interest")
plt.plot(month_number,principal_paid,c='b',label="Principal")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(left = 1,right=70)
plt.ylim(bottom = 0,top = 1000)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Dollars",fontsize=20)
plt.title("Interest and Principal paid each month",fontsize=25)
plt.legend(loc=(1.02,0),borderaxespad=0,fontsize=20)

plt.savefig('images/interest_principal.png',dpi=300)
```

```python
plt.figure(figsize=(10,5))
plt.plot(month_number,interest_paid,c='k',label="Interest")
plt.plot(month_number,principal_paid,c='b',label="Principal")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(left = 1,right=70)
plt.ylim(bottom = 0,top = 1000)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Dollars",fontsize=20)
plt.title("Interest and Principal paid each month",fontsize=25)
plt.legend(loc=(1.02,0),borderaxespad=0,fontsize=20)

plt.tight_layout()
plt.savefig('images/interest_principal_fit.png',dpi=300)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(10,5))
axes.plot(month_number,interest_paid,c='k',label="Interest")
axes.plot(month_number,principal_paid,c='b',label="Principal")
axes.tick_params(axis='x',labelsize=10)
axes.tick_params(axis='y',labelsize=10)
axes.set_xlim(left = 1,right=70)
axes.set_ylim(bottom = 0,top = 1000)
axes.set_xlabel("Month",fontsize=20)
axes.set_ylabel("Dollars",fontsize=20)
axes.set_title("Interest and Principal paid each month",fontsize=25)
axes.legend(loc=(1.02,0),borderaxespad=0,fontsize=20)

fig.savefig('images/interest_principal_object-oriented.png',dpi=300)
```
